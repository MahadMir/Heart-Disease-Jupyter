```python
#Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
from sklearn import tree
from sklearn import svm
from sklearn.cluster import KMeans
```

```python
#Loading the dataframe

diseaseDataFrame = pd.read_csv("heart.csv")
```

```python
#Dropping Target

predictionData = diseaseDataFrame.target.values
predictorData = diseaseDataFrame.drop(columns ='target', axis = 1)
```

```python
#making a dataframe of y values

predictionDataFrame = pd.DataFrame(predictionData)
predictorDataFrame = pd.DataFrame(predictorData)

print(predictionDataFrame.head())
print("\n",predictorDataFrame.head())
```

```python
x = predictorDataFrame.values
y = predictionDataFrame.values
```

```python
#Splitting the Dataset into 80/20 train/test

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
#K Nearest Neighbours
def nearestNeighbourHeartDisease(i):
    neighboursHeartDisease = KNeighborsClassifier(n_neighbors=i)

    neighboursHeartDisease.fit(x_train, y_train.ravel())

    neighboursHeartDiseasePrediction = neighboursHeartDisease.predict(x_test)

    neighboursHeartDiseasePredictionScore = neighboursHeartDisease.score(x_test, y_test.ravel())*100
    return neighboursHeartDiseasePredictionScore
    
print(nearestNeighbourHeartDisease(2))
```

```python
listNeighbourHeartDisease = []
for i in range(20):
    listNeighbourHeartDisease.append(nearestNeighbourHeartDisease(i+1))

xAxis = range(1, 21)
yAxis = listNeighbourHeartDisease
plt.plot(xAxis, yAxis)
```

```python
#Decision Tree

decisionTreeHeartDisease = DecisionTreeClassifier(criterion = 'entropy')
decisionTreeHeartDisease.fit(x_train, y_train)

decisionTreeHeartDiseasePredictionScore = decisionTreeHeartDisease.score(x_test, y_test)*100

print(decisionTreeHeartDiseasePredictionScore)
```

```python
plt.figure(figsize=(40,40))
x = tree.plot_tree(decisionTreeHeartDisease)
plt.show()
```

```python
#Naive Bayes

naiveBayesHeartDisease = GaussianNB()
naiveBayesHeartDisease.fit(x_train, y_train.ravel())

naiveBayesHeartDiseasePredictionScore = naiveBayesHeartDisease.score(x_test, y_test.ravel())*100

print(naiveBayesHeartDiseasePredictionScore)
```

```python
#Linear Regression

linearRegressionHeartDiseaese = LinearRegression()
linearRegressionHeartDiseaese.fit(x_train, y_train)

linearRegressionHeartDiseaesePredictionScore = linearRegressionHeartDiseaese.score(x_test, y_test)*100
print(linearRegressionHeartDiseaesePredictionScore)
```

```python
#SVM

SVMHeartDisease = svm.SVC()
SVMHeartDisease.fit(x_train, y_train.ravel())

SVMHeartDiseaesePredictionScore = SVMHeartDisease.score(x_test, y_test)*100
print(SVMHeartDiseaesePredictionScore)
```
